```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('Titanic.csv')
```

```python
df
```

```python
# Apply is one of the most powerful function in Pandas that helps to manipulate data in data table by own rule
```

```python
def get_Class(cls):
    if cls=='A':
        return "LuxuryClass"
    elif cls=='B':
        return "RichClass"
    elif cls== 'C':
        return "UpperClass"
    else:
        return "MiddleClass"
```

```python
df.Class= df.Class.apply(get_Class)
```
