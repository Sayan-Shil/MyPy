```python
import numpy as np
import pandas as pd
```

```python
columns = ['Item', 'Price', 'Quantity', 'Revenue', 'Discount', 'Net Profit']
data = {
    'Item': ['Apple', 'Banana', 'Carrot', 'Dates', 'Eggplant', 'Fig', 'Grapes', 'Honeydew', 
             'Ivy Gourd', 'Jackfruit', 'Kiwi', 'Lemon', 'Mango', 'Nectarine', 'Orange', 
             'Papaya', 'Quince', 'Raspberry', 'Strawberry', 'Tomato'],
    'Price': np.random.uniform(10, 100, 20),
    'Quantity': np.random.randint(1, 50, 20),
    'Revenue': np.random.uniform(500, 5000, 20),
    'Discount': np.random.uniform(0, 30, 20),
    'Net Profit': np.random.uniform(50, 1000, 20)
}

df = pd.DataFrame(data)
df
```

```python
df.plot(kind="line")
```
